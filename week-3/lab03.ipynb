```python
# A
"""
File: encrypt.py
Encrypts an input string of lowercase letters and prints
the result. The other input is the distance value.
"""
plainText = input("Enter a one-word, lowercase message: ")
distance = int(input("Enter the distance value: "))
code = " "
for ch in plainText:
  ordvalue = ord(ch)
  cipherValue = ordvalue + distance
  if cipherValue > ord('z'):
    cipherValue = ord('a') + distance - (ord('z') - ord(ch) + 1)
  code += chr(cipherValue)
print(code)

```

```python
# A
"""
File: decrypt.py
Decrypts an input string of lowercase letters and prints
the result. The other input is the distance value.
"""
code = input("Enter the coded text: ")
distance = int(input("Enter the distance value: "))
plainText = " "
for ch in code:
  ordvalue = ord(ch)
  cipherValue = ordvalue - distance
  if cipherValue < ord('a'):
    cipherValue = ord('z') - abs(distance - (ord('a') - ordvalue - 1))
  plainText += chr(cipherValue)
print(plainText)
```

```python
# B

def vowel_detector(string):
  vowels = "aeiou"
  vow_rem = []
  count = 0
  for letter in string:
    if letter not in vowels:
      vow_rem.append(letter)
      count += 1
  return vow_rem, count

string = input("Enter a string: ")
converted, count = vowel_detector(string)
output = "".join(converted)
print(output, count)
```

```python
# D-5
"""
A bit shift is a procedure whereby the bits in a bit string are moved to the left or
to the right. For example, we can shift the bits in the string 1011 two places to the
left to produce the string 1110. Note that the leftmost two bits are wrapped
around to the right side of the string in this operation. Define two scripts,
shiftLeft.py and shiftRight.py, that expect a bit string as an input. The script
shiftLeft shifts the bits in its input one place to the left, wrapping the leftmost bit
to the rightmost position. The script shiftRight performs the inverse operation.
Each script prints the resulting string.
"""
def shiftLeft(bits, distance):
    shifted_bits = list(bits)
    distance = int(distance)

    tmp = shifted_bits[len(shifted_bits) - 1 - distance]
    shifted_bits[len(shifted_bits) - 1 - distance] = shifted_bits[-1]
    shifted_bits[-1] = tmp

    return shifted_bits


def shiftRight(bits, distance):
    shifted_bits = list(bits)
    distance = int(distance)

    tmp = shifted_bits[distance]
    shifted_bits[distance] = shifted_bits[0]
    shifted_bits[0] = tmp

    return shifted_bits


bits = input("Enter a binary number: ")
distance = input("Enter the distance you want the the rightmost bits to move left: ")
li = shiftLeft(bits, distance)
text = "".join(li)
print(text)

li = shiftRight(bits, distance)
text = "".join(li)
print(text)
```

```python
# D-6
"""
Use the strategy of the decimal to binary conversion and the bit shift left operation
defined in Project 5 to code a new encryption algorithm. The algorithm should add
1 to each character’s numeric ASCII value, convert it to a bit string, and shift the
bits of this string one place to the left. A single-space character in the encrypted
string separates the resulting bit strings
"""
def addOneASCII(plainText):
    for ch in plainText:
        ordvalue = ord(ch)
        cipherValue = ordvalue + 1
    return(cipherValue)


def decToBinary(number):
    if number == 0:
        return 0

    bitString = ""

    while number > 0:
        remainder = number % 2
        number = number // 2
        bitString = str(remainder) + bitString

    return bitString


def shiftLeft(bits):
    shifted_bits = list(bits)

    tmp = shifted_bits[len(shifted_bits) - 2]
    shifted_bits[len(shifted_bits) - 2] = shifted_bits[-1]
    shifted_bits[-1] = tmp

    return shifted_bits


text = input("Enter a string: ")
ascii_number = addOneASCII(text)
print(ascii_number)

binary = decToBinary(ascii_number)
print(binary)

shifted = shiftLeft(binary)
print(shifted)

shifted_text = " ".join(shifted)
print(shifted_text)
```
