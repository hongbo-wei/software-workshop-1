
Q4. Modify the Counter class further so that each counter can have its own individual (and different) 'wrap around' point. 

```python
class Counter:

    def __init__(self, wrap_around):
        self.theCount = 0
        self.wrap_aroud = wrap_around

    def inc(self):
        if self.theCount < self.wrap_aroud:
            self.theCount = self.theCount + 1
        else:
            self.theCount = 0

    def dec(self):
        if self.theCount > 0:
            self.theCount = self.theCount - 1
        else:
            self.theCount = self.wrap_aroud

    def getCount(self):
        return self.theCount

    def setCount(self, newCount):
        self.theCount = newCount

    def equals(self, other):
        return self.theCount == other.theCount
    
    # Q2
    def update(self, qty):
        self.theCount += qty
        if self.theCount > self.wrap_aroud:
            self.theCount = self.theCount % (self.wrap_aroud + 1)


X = "aabbcc"

A = Counter (999)
B = Counter (499)
C = Counter (4999)
D = C
E = B

A.inc()
for char in X :
    A.inc()

B.dec()
B.dec()

B.dec()
B.inc()

C.inc()
C.setCount(7)
print(A.getCount(), B.getCount(), E.getCount(), C.getCount(), \
      C==D, A==C, A.equals(C))
```


Q7. Add a method called encypher that takes a plain text character string as a parameter instead of a single character, and returns the encyphered version of the whole character string . Add another method called decypher that takes a cypher text string and returns the plain text version of the string.

```python
class Caesar():
    # alphabet = "abcdefghijklmnopqrstuvwxyz .,"
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz .,"

    def __init__(self, shift):
        self.shift = shift

    def encode(self, char):
        position = Caesar.alphabet.index(char)
        newpos = position + self.shift

        if newpos >= len(Caesar.alphabet):
            newpos = newpos - len(Caesar.alphabet)

        newchar = Caesar.alphabet[newpos]
        return newchar

    def decode(self, char):
        position = Caesar.alphabet.index(char)
        newpos = position - self.shift

        if newpos < 0:
            newpos = newpos + len(Caesar.alphabet)

        newchar = Caesar.alphabet[newpos]
        return newchar
    
    def encypher(self, string):
        encyphered_text = ''
        for char in string:
            encyphered_text += self.encode(char)
        return encyphered_text
    
    def decypher(self, string):
        decyphered_text = ''
        for char in string:
            decyphered_text += self.decode(char)
        return decyphered_text

machine = Caesar (4)

cyphertext = "lipps"

enciphered_text = machine.encypher(cyphertext)
print(enciphered_text)

deciphered_text = machine.decypher(cyphertext)
print(deciphered_text)
```

```python
class BouncyBall:
    def __init__(self, price, size, brand):
        self.__price = 0 
        self.__size = ''
        self.__brand = ''

        # Set initial values using setters to enforce constraints
        self.set_price(price)
        self.set_size(size)
        self.set_brand(brand)

    def set_price(self, price):
        if price <= 0:
            print("Price must be greater than 0. Setting price to 0.")
            self.__price = 0
        else:
            self.__price = price
        
    def set_size(self, size):
        valiad_size = ['small', 'medium', 'large']
        if size.lower() not in valiad_size:
            print('Invalid Size, choose from small, medium, and large. Setting size to empty string.')
            self.__size = ''
        else:
            self.__size = size.lower()
    
    def set_brand(self, brand):
        if isinstance(brand, str):
            self.__brand = brand
        else:
            print('Invalid input. Setting brand to empty string.')
            self.__brand = ''

    def get_price(self):
        return self.__price
    
    def get_size(self):
        return self.__size
    
    def get_brand(self):
        return self.__brand
    

boxing_ball = BouncyBall(99, 'small', 'Dragon')
print(boxing_ball.get_price())
print(boxing_ball.get_size())
print(boxing_ball.get_brand())

badminton_ball = BouncyBall(-5, 'XL', 99)
print(badminton_ball.get_price())
print(badminton_ball.get_size())
print(badminton_ball.get_brand())
```
