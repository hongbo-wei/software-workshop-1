```python
# 1
import math

def newton(number):
    '''
    Package Newton's way of calculate sqaure root
    '''
    x = number
    tolerance = 1e-6
    estimate = 1

    while True:
        estimate = (estimate + x / estimate) / 2
        difference = abs(x - estimate ** 2)
        if difference <= tolerance:
            break

    return estimate


def main():
    user_input = input("Enter a positive number or press 'Enter' to quit: ")
    if user_input == "":
        return
    else:
        x = float(user_input)

    print(f"The program's estimate:", newton(x))
    print(f"Python's estimate:", math.sqrt(x))

if __name__ == '__main__':
    main()
```

```python
# 2
def isSorted(list):
    if len(list) == 1:
        return True
    else:
        count = 1
        for i in range(len(list)):
            if list[i] > list[count]:
                return False
                break
            count += 1
            if count > len(list) - 1:
                break
        return True


li = [2, 4, 5, 6]
isSorted(li)
```
